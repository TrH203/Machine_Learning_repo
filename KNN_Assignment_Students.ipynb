2. Viết hàm tính khoảng cách từ điểm P đếm M lưu kết quả tính được vào cột số 0 vừa thêm vào

```python
def dist_ps_fast(P, M):
    M_new = np.sum(M*M,1) 
    z_new = np.sum(P*P) 
    
    return np.sqrt(M_new + z_new - 2*M.dot(P))
```

3. Viết hàm sắp xếp M theo thứ tự tăng dần của cột đầu tiên (cột vừa tính được từ câu 2).

```python
def sap_xep(M, col=0):
    for i in range(len(M) - 1):
        for j in range(i+1,len(M)):
            #print(M[i][col] , M[j][col])
            if M[i][col] > M[j][col]:
                M[[i, j]] = M[[j,i]]
    return M
```

12. Tổ chức các hàm đã viết thành lớp lấy tên là KNN_Ten_Ban

```python
class KNN_Hien():
    def __init__(self,x_train,y_train) -> None:
        self.pre_process(x_train=x_train,y_train=y_train)
        
        
    def pre_process(self,x_train,y_train):
        self.XX = np.hstack((x_train,y_train.reshape(-1,1)))
        self.XX = np.hstack((np.zeros((self.XX.shape[0],1)), self.XX))
        #print(self.XX)
    
    def fit(self, test_data):
        self.ds = []
        if test_data.shape[0] > 1:
            tmp = self.dist_cal(test_data,self.XX[:,1:-1])
            for i,j in zip(tmp,range(len(tmp))):
                xx = np.copy(self.XX)
                xx[:,0] += i
                self.ds.append(self.sap_xep(xx))
                #print(self.sap_xep(xx))
            #print(self.ds[8])
                       
        else:
            xx = np.copy(self.XX)
            xx[:,0] += self.dist_1point_fast(i,xx[:,1:-1])
            self.ds.append(self.sap_xep(xx))

    
            
        
    def sap_xep(self,M, col=0):
        for i in range(len(M) - 1):
            for j in range(i+1,len(M)):
                #print(M[i][col] , M[j][col])
                if M[i][col] > M[j][col]:
                    M[[i, j]] = M[[j,i]]
        return M
    
    
    def dist_1point_fast(self,P, M):
        M_new = np.sum(M*M,1) 
        z_new = np.sum(P*P) 
        return np.sqrt(M_new + z_new - 2*M.dot(P))
    
    def dist_Npoint_fast(self,P,M):
        M_new = np.sum(M*M, 1) 
        P_new = np.sum(P*P, 1) 
        return np.sqrt(P_new.reshape(-1,1) + M_new.reshape(1,-1) - 2*P.dot(M.T))

    def dist_cal(self, P,M):
        try:
            rs = self.dist_1point_fast(P,M)
        except:
            rs = self.dist_Npoint_fast(P,M)
            
        return rs
            
    # list1 = M[:3,M.shape[1]-1].tolist()
    def counter_words_in_list(self,llist):
        d = {}
        for i in llist:
            if d.get(i):
                d[i] += 1
            else:
                d[i] = 1
        m = 0
        word = ""
        for i in d.keys():
            if m < d[i]:
                m = d[i]
                word = i
        return [word, m]

    def predict(self,k=5):
        rs = []  
        for i in self.ds:
            rs.append(self.counter_words_in_list(i[:k,-1])[0])
        return rs
```
