1. Viết lại hàm tách dữ liệu thành tập Train và tập Test

```python
import numpy as np
import pandas as pd
```

```python
def train_test_split(dataset, test_Number):
  import random
  import numpy as np
    #khởi tạo tập test rỗng
  test_data= []
    #lấy từng hàng ở tập dataset đưa vào tập test
  while len(test_data) < test_Number:
        #lấy index ngẫu nhiên trong tập tập dữ liệu
      index = random.randrange(len(dataset))
        #đưa hàng thứ index vào tập test đồng thời xóa ở dataset
        #chuyển dataset qua list để dùng hàm pop. Hàm này dùng để xóa hàng
      test_data.append(list(dataset).pop(index))
  temp=test_data
  y_test=np.array(temp)[:,-1]
    #y_test=X_test[:,-1]
   # X_test1=X_test[:,:-1]
  return dataset,test_data,y_test
```

2. Viết hàm tính kỳ vòng. Kỳ vòng được tính bằng 1/n*sum(xi)

```python
def mean(numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / float(len(numbers))
```

3. Viết hàm tính phương sai. Phương sai được tính bằng căn bậc hai của:
# 1/(n-1)*sum(bình phương(xi-kỳ vọng))

```python
import math
def standard_deviation(numbers):
    if len(numbers) < 2:
        return 0
    avg = mean(numbers)
    variance = sum([(x - avg)**2 for x in numbers]) / float(len(numbers) - 1)
    return math.sqrt(variance)
```

```python
a = np.array([[1,2,3,4],
              [4,3,2,1]])
print(a.shape)
standard_deviation(a[:,0])
```

4. Tính xác suất phân phối Gaussian, dựa trên độ lệch chuẩn và phương sai

```python
def summarize(dataset):
    dataset=np.array(dataset)
    temp=np.zeros((dataset.shape[1],2))
    print(temp.shape)
    for i in range(dataset.shape[1]):
        temp[i,0]=mean(dataset[:,i])
        temp[i,1]=standard_deviation(dataset[:,i])
    return temp
```

```python
import math
def calculate_prob(x, mean, stdev):
    exponent = np.exp(-(x - mean)**2 / (2 * stdev*stdev))
    return (1 / (np.sqrt(2 * np.pi) * stdev*stdev)) * exponent
```

```python
def get_accuracy(value, y_test):
    correct = 0
    for i in range(len(value)):
        if value[i] == y_test[i]:
            correct += 1
    return (correct / float(len(value))) * 100.0
```
